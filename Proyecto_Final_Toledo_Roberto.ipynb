## ⏩ Eliminación de outliers

```python
def limpiar_datos_IQR(columna,considerar):
    print("\n\t---- {} ----".format(columna))

    q1 = considerar[columna].quantile(0.25) #Calculamos el 1 quartil
    q3 = considerar[columna].quantile(0.75) # calculamos el 3 cuartil
    iqr = q3 - q1 # calculamosel IQR
    print("Q1 = {:,.4f}\nQ3 = {:,.4f} \nIQR = {:,.4f}".format(q1,q3,iqr))

    # Definición de los límites superior e inferior
    limite_inferior = q1 - (iqr * 1.5)
    limite_superior = q3 + (iqr * 1.5)
    print("IQR x 1.5 = {:,.4f}\nLímite inferior = {:,.4f}\nLímite superior = {:,.4f}".format((iqr*1.5), limite_inferior, limite_superior))
    return considerar [(considerar[columna]>limite_inferior) & (considerar[columna]<limite_superior)]

df = limpiar_datos_IQR("colesterol",df)
df = limpiar_datos_IQR("trigliceridos",df)
df = limpiar_datos_IQR("edad",df)
```

## ⏩ Hipótesis inicial (H1)

### "En el dataset analizado, la población masculina presenta una mayor incidencia de infartos en comparación con la femenina. Esto sugiere que los hombres podrían tener un mayor riesgo de sufrir un infarto, lo que podría estar influenciado por factores como diferencias biológicas, hormonales o de estilo de vida."

```python
plt.figure(figsize=(8, 6))  # Aumentamos el tamaño para más claridad

# Crear el gráfico de barras vertical
ax = sns.countplot(x="sexo", data=df, palette="viridis", hue="riesgo_infarto")

# Obtener el total de registros
total = len(df)
max_value = df["sexo"].value_counts().max()  # Valor máximo para ajustar el límite

# Agregar etiquetas con cantidad y porcentaje
for p in ax.patches:
    count = int(p.get_height())  # Cantidad
    percentage = 100 * count / total  # Porcentaje

    ax.annotate(f'{count}\n({percentage:.1f}%)', (p.get_x() + p.get_width() / 2, p.get_height() + max_value * 0.02),
                ha='center',
                va='bottom',
                fontsize=10,
                color='black',
                bbox=dict(
                    boxstyle="round,pad=0.3",
                    fc="white",
                    ec="black",
                    lw=0.5))

# Ajustar el límite del eje Y para dar más espacio
plt.ylim(0, max_value * 1.15)

# Mejorar distribución del gráfico
plt.title("Distribución del riesgo de infarto por sexo")
plt.xlabel("Sexo")
plt.ylabel("Cantidad")
plt.tight_layout()
plt.show()
```

### ⏩ A simple vista el gráfico nos sugiere que el Hombre es mas propenso a sufrir un infarto que las mujeres. No obstante realizaremos la prueba de Chi-Cuadrado para verificar si existe una diferencia significativa en la frecuencia de infartos entre hombres y mujeres

Hipótesis Estadísticas:

Hipótesis Nula (
𝐻
0
H
0
​
 ): No hay diferencia significativa en la proporción de infartos entre hombres y mujeres (las diferencias observadas se deben al azar).

Hipótesis Alternativa (
𝐻
1
H
1
​
 ): Sí hay una diferencia significativa en la proporción de infartos entre hombres y mujeres.

```python

from scipy.stats import chi2_contingency

# 🔹 2️⃣ Revisar los valores únicos en la columna de género
print(df['sexo'].value_counts())

# 🔹 3️⃣ Contar los casos de infarto por género
tabla_contingencia = pd.crosstab(df["sexo"], df["riesgo_infarto"])  # Infarto: 1 = Sí, 0 = No
print("Tabla de Contingencia:\n", tabla_contingencia)

# 🔹 4️⃣ Calcular la tasa de infartos por género
tasa_infartos = tabla_contingencia.div(tabla_contingencia.sum(axis=1), axis=0) * 100
print("\nTasa de Infartos por Género (%):\n", tasa_infartos)

# 🔹 5️⃣ Prueba de Chi-Cuadrado
chi2, p, dof, expected = chi2_contingency(tabla_contingencia)
print(f"\nResultados de la Prueba de Chi-Cuadrado:")
print(f"Chi2 = {chi2:.4f}")
print(f"p-valor = {p:.4f}")

# 🔹 6️⃣ Interpretación del resultado
alpha = 0.05  # Nivel de significancia
if p < alpha:
    print("\n✅ Rechazamos la hipótesis nula: Hay una diferencia significativa en la tasa de infartos entre hombres y mujeres.")
else:
    print("\n❌ No podemos rechazar la hipótesis nula: No hay suficiente evidencia para afirmar que hay una diferencia significativa.")
```

### Conclusión: No hay evidencia suficiente para afirmar que hay una diferencia significativa en la tasa de infartos entre hombres y mujeres.

```python
# Contar la cantidad de pacientes en cada categoría de presión arterial
conteo_presion = df["tipo_presion_arterial"].value_counts()

# Calcular la tasa de infartos por categoría de presión arterial
tasa_infartos = df.groupby("tipo_presion_arterial")["riesgo_infarto"].mean() * 100

# Visualización
plt.figure(figsize=(8,5))
tasa_infartos.sort_values().plot(kind="bar", color="skyblue", edgecolor="black")
plt.xlabel("Categoría de Presión Arterial")
plt.ylabel("Tasa de Infartos (%)")
plt.title("Tasa de Infartos por Nivel de Presión Arterial")
plt.xticks(rotation=45)
plt.show()
```
