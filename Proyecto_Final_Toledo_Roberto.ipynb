## ⏩ Eliminación de outliers

```python
def limpiar_datos_IQR(columna,considerar):
    print("\n\t---- {} ----".format(columna))

    q1 = considerar[columna].quantile(0.25) #Calculamos el 1 quartil
    q3 = considerar[columna].quantile(0.75) # calculamos el 3 cuartil
    iqr = q3 - q1 # calculamosel IQR
    print("Q1 = {:,.4f}\nQ3 = {:,.4f} \nIQR = {:,.4f}".format(q1,q3,iqr))

    # Definición de los límites superior e inferior
    limite_inferior = q1 - (iqr * 1.5)
    limite_superior = q3 + (iqr * 1.5)
    print("IQR x 1.5 = {:,.4f}\nLímite inferior = {:,.4f}\nLímite superior = {:,.4f}".format((iqr*1.5), limite_inferior, limite_superior))
    return considerar [(considerar[columna]>limite_inferior) & (considerar[columna]<limite_superior)]

df = limpiar_datos_IQR("colesterol",df)
df = limpiar_datos_IQR("trigliceridos",df)
df = limpiar_datos_IQR("edad",df)
```

# ▶️ Feature engineering (Ingenieria de caracteristicas)
<p>Es el proceso de crear nuevas características a partir de las existentes, utilizando nuestro conocimiento del dominio y los datos. Algunas técnicas comunes incluyen:

Transformación de variables (logarítmica, cuadrática, etc.)
Creación de características mediante operaciones entre variables existentes
Codificación de variables categóricas
Normalización y estandarización
Manejo de valores atípicos y faltantes</p>

```python
# creación de diccionario para los tipos de colesterol LDL segun su valor
clasificacion_colesterol_LDL = {
    "optimo":(0,100),
    "mejor que optimo":(100,129),
    "limite alto":(130,159),
    "alto":(160,189),
    "muy alto":(190,float("inf"))
}
# creación de diccionario para los tipos de trigliceridos segun su valor
clasificacion_trigliceridos = {
    "saludables":(0,150),
    "ligeramente altos":(150,190),
    "altos":(200,499),
    "muy altos":(500,float("inf"))
}
# funcion que retorna el tipo de colesterol
def categorizar_colesterol(valor):
    for categoria, (inferior,superior) in clasificacion_colesterol_LDL.items():
        if inferior <= valor <= superior:
            return categoria

# funcion que retorna el tipo de presión
def categorizar_presion(sis,dia):
    if sis <120 and dia <80:
        return "normal"
    elif 120 <= sis <=129 and dia < 80:
        return "elevada"
    elif (130 <= sis <=139) or (80 <= dia <= 89):
        return "hipertension grado 1"
    elif (140 <= sis <= 179) or (90 <= dia <=119):
        return "hipertension grado 2"
    elif sis >=180 or dia >=120:
        return "crisis hipertensiva"
    else:
        return "no clasificado"

def categorizar_trigliceridos(valor):
    for categoria, (inferior, superior) in clasificacion_trigliceridos.items():
        if inferior <= valor <=superior:
            return categoria

 # generación de columna de tipo de colesterol LDL
df["clasificacionLDL"] = df["colesterol"].apply(categorizar_colesterol)

# Generación de columna clasificación de trigliceridos
df["clasificacion_trigliceridos"] = df["trigliceridos"].apply(categorizar_trigliceridos)

# Cortamos la columna mediante el separador / y las dividimos dos columnas separadas
df[["sistolica","diastolica"]] = df["presion_arterial"].str.split("/",expand=True)
# Convertimos las columnas a numero
df["sistolica"] = pd.to_numeric(df["sistolica"])
df["diastolica"] = pd.to_numeric(df["diastolica"])
# llamos a la funcion que genera la columna categorica en funcion de los valores de la columnas anteriormente generadas
df["tipo_presion_arterial"] = df.apply(lambda row: categorizar_presion(row["sistolica"],row["diastolica"]),axis=1)

# visualizamos las columnas nuevas generadas junto con las originales
df[["presion_arterial","sistolica","diastolica","tipo_presion_arterial","colesterol","clasificacionLDL","trigliceridos","clasificacion_trigliceridos"]]
```

## ⏩ Preparación de los datos
### Asignación de pesos inversamente proporcionales a la frecuencia de clases, para corregir el desbalance de la clase 0 von la clase 1

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# Variable predictora (X) y objetivo (y)
X = df[["colesterol"]]  # Variable numérica continua
y = df["riesgo_infarto"]  # Variable binaria (0: no riesgo, 1: riesgo)

# Modelo con balanceo de clases
model = LogisticRegression(class_weight='balanced', random_state=42)
model.fit(X, y)
```

## ⏩ Evaluación del modelo

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X)
print("Matriz de Confusión:")
print(confusion_matrix(y, y_pred))
print("\nReporte de Clasificación:")
print(classification_report(y, y_pred))
```

### Interpretación
#### Recall (Sensibilidad) de clase 1 = 50%: El modelo identifica correctamente el 50% de los pacientes con riesgo.

#### Precisión de clase 1 = 45%: De los predichos como riesgo, el 45% realmente lo tiene.

## Curva de ROC y AUC

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

y_proba = model.predict_proba(X)[:, 1]
auc = roc_auc_score(y, y_proba)

fpr, tpr, _ = roc_curve(y, y_proba)
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("Tasa de Falsos Positivos")
plt.ylabel("Tasa de Verdaderos Positivos")
plt.legend()
plt.show()
```

### Interpretación
#### AUC ≈ 0.51: el modelo no es mejor que el aleatorio

#### Justificación:

#### obesidad binaria: Facilita la interpretación en modelos estadísticos.

#### Variable compuesta: Captura la interacción entre sedentarismo, bajo ejercicio y obesidad.

```python
#. Creación de variable compuesta "estilo_de_vida_riesgo" (combina sedentarismo, bajo ejercicio y obesidad)
df['estilo_de_vida_riesgo'] = (
    (df['horas_sedentarismo_diarias'] > df['horas_sedentarismo_diarias'].median()) &  # Alto sedentarismo
    (df['horas_ejercicio_semanal'] < df['horas_ejercicio_semanal'].median()) &       # Bajo ejercicio
    (df['obesidad'] == 1)                                                            # Obesidad
).astype(int)  # 1: Cumple criterios de riesgo, 0: No cumple
```
